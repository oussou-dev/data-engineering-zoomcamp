### Data Schema

```python
import pandas as pd

# use fastparquet engine to avoid PyArrow extension errors in this environment
df = pd.read_parquet("data/green_tripdata_2025-11.parquet", engine="fastparquet")

# take first 100 rows
df = df.head(100)


# convert datetime
df["lpep_pickup_datetime"] = pd.to_datetime(df["lpep_pickup_datetime"])
df["lpep_dropoff_datetime"] = pd.to_datetime(df["lpep_dropoff_datetime"])


parse_dates = [
    "lpep_pickup_datetime",
    "lpep_dropoff_datetime"
]

# enforce dtypes after reading
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64",
}
df = df.astype(dtype)

df.head()
```

```python
from sqlalchemy import create_engine
import pyarrow.parquet as pq
from tqdm.auto import tqdm

parquet_file = pq.ParquetFile("data/green_tripdata_2025-11.parquet")

for batch in tqdm(
    parquet_file.iter_batches(batch_size=100_000),
    total=parquet_file.metadata.num_rows // 100_000 + 1,
    desc="Ingesting parquet"
):
    df = batch.to_pandas()
    df = df.astype(dtype)

    df.to_sql(
        name="green_tripdata_2025_11",
        con=engine,
        if_exists="append",
        index=False,
        method="multi"
    )
```

```python
# Read the data
green_taxi_trips = pd.read_parquet("./data/green_tripdata_2025-11.parquet", engine = "fastparquet")
zone_lkup = pd.read_csv("./data/taxi_zone_lookup.csv")

display(green_taxi_trips.head(10), zone_lkup.head(10))
```

```python
# Merge the data sets for easier reference

# Pick Up Location
trips_and_zones = green_taxi_trips.merge(
    zone_lkup,
    left_on = "PULocationID",
    right_on = "LocationID",
    how = "left")

trips_and_zones = trips_and_zones.rename(
    columns={
        "Borough": "Borough_PU",
        "Zone": "Zone_PU",
        "service_zone": "service_zone_PU"
    }
)

trips_and_zones = trips_and_zones.drop(columns=["LocationID"])

# Drop Off Location
trips_and_zones = trips_and_zones.merge(
zone_lkup,
left_on = "DOLocationID",
right_on = "LocationID",
how = "left")

trips_and_zones = trips_and_zones.rename(
    columns={
        "Borough": "Borough_DO",
        "Zone": "Zone_DO",
        "service_zone": "service_zone_DO"
    }
)

trips_and_zones = trips_and_zones.drop(columns=["LocationID"])

trips_and_zones.head(10)
```

### Question 3. Counting short trips

```python
november_trips = green_taxi_trips[
    (green_taxi_trips['lpep_pickup_datetime'] >= '2025-11-01') &
    (green_taxi_trips['lpep_pickup_datetime'] < '2025-12-01') &
    (green_taxi_trips['trip_distance'] <= 1)
]

short_trips_count = len(november_trips)
short_trips_count
```

### Question 4. Longest trip for each day

```python
# Extract the date from lpep_pickup_datetime and find the longest trip for each day
longest_trip_per_day = green_taxi_trips[green_taxi_trips['trip_distance'] < 100].groupby(
    green_taxi_trips['lpep_pickup_datetime'].dt.date
)['trip_distance'].max()

# Find the day with the longest trip
day_with_longest_trip = longest_trip_per_day.idxmax()
longest_distance = longest_trip_per_day.max()

print(f"Day with longest trip: {day_with_longest_trip}")
print(f"Longest trip distance: {longest_distance} miles")
```

### Question 5. Biggest pickup zone

```python
# Filter trips for November 18th, 2025
nov_18_trips = trips_and_zones[
    trips_and_zones['lpep_pickup_datetime'].dt.date == pd.to_datetime('2025-11-18').date()
]

# Group by Zone and sum the total_amount
pickup_zone_totals = nov_18_trips.groupby('Zone_PU')['total_amount'].sum()

# Find the pickup zone with the largest total_amount
biggest_pickup_zone = pickup_zone_totals.idxmax()
largest_total = pickup_zone_totals.max()

print(f"Pickup zone with largest total_amount on Nov 18: {biggest_pickup_zone}")
print(f"Total amount: ${largest_total:.2f}")
```

### Question 6. Largest tip

```python
# Filter trips with pickup zone "East Harlem North" in November 2025
east_harlem_north_pickups = trips_and_zones[
    (trips_and_zones['Zone_PU'] == 'East Harlem North') &
    (trips_and_zones['lpep_pickup_datetime'].dt.month == 11) &
    (trips_and_zones['lpep_pickup_datetime'].dt.year == 2025)
]

east_harlem_north_pickups = east_harlem_north_pickups.groupby(
    trips_and_zones['Zone_DO'])['tip_amount'].max()

zone = east_harlem_north_pickups.idxmax()
amt = east_harlem_north_pickups.max()

print(f"Dropoff zone with largest tip: {zone}")
print(f"Tip amount: ${amt:.2f}")
```
